## Simple RatInABox 
Initialise a 2D Environment. Initialise an Agent in the environment. Initialise some PlaceCells. Simulate for 60 seconds. Print table of times, position and firing rates and spikes. Plot trajectory and rate timeseries'.

```python
# Import ratinabox 
import ratinabox
from ratinabox.Environment import Environment
from ratinabox.Agent import Agent
from ratinabox.Neurons import *
import pandas as pd 
```

```python
# Run a very simple simulation 
Env = Environment()
Ag = Agent(Env)
PCs = PlaceCells(Ag)
for i in range(int(60 / Ag.dt)):
    Ag.update()
    PCs.update()
```

```python
# Export data into a dataframe 
pd.DataFrame(Ag.history)
```

```python
# Plot data 
Ag.plot_trajectory()
PCs.plot_rate_timeseries()
```
